#### 프롬프트
- 모델이 무엇을 해야 하는지 설명하는 텍스트 input

#### 프롬프트 엔지니어링
- 모델과 대화를 잘 하는 방법
- 모델이 최대한 일관되게, 내가 원하는 결과를 내도록 지시문을 설계하는 과정
- 같은 프롬프트, 같은 모델, 같은 환경이어도 결과가 100% 동일하지 않을 수 있음

```python
response
```

- reasoning : 모델의 내부 추론 강도 조절 (출력 속도)
    - low : 빠르게, 직관적으로
    - medium : 일반적인 설명
    - high : 복잡한 문제를 깊게 추론
- instructions : 모델에게 지시하는 부분 (최상위 규칙)

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    reasoning={"effort": "low"},
    instructions="Talk like a pirate.",
    input="Are semicolons optional in JavaScript?",
)

print(response.output_text)
```

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-nano",
    instructions="""
    You are a backend API assistant.
    - Answer concisely and accurately.
    - If unsure, say you are unsure.
    - Do not hallucinate APIs or facts.
    - Prefer bullet points over long prose.
    """,
    input="Are semicolons optional in JavaScript?",
)

print(response.output_text)
```

- You are a backend API assistant. : 모델 정체성 지정
- Answer concisely and accurately. : 짧고, 핵심위주로, 정확하게
- If unsure, say you are unsure. : 모르면 모른다고 하고
- Do not hallucinate APIs or facts. : 환각 활률 낮추기
- Prefer bullet points over long prose. : 목록 형태로

#### 역할 부여
- developer
- user : input 메세지 대상자
- assistant : 모델이 생성한 답변을 가지고 있는 역할

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-mini",
    input=[
        {
            "role":"developer",
            "content":"너는 월드컵에 대해 많은 정보를 가지고 있는 축구 기자야"
        },
        {
            "role":"user",
            "content":"2002년 월드컵에서 가장 화제가 되었던 나라는 어느 곳이야?"
        }
    ]
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-5-mini",
    input=[
        {
            "role":"user",
            "content":"2002년 월드컵에서 가장 화제가 되었던 나라는 어느 곳이야?"
        },
        {
            "role":"assistant",
            "content":"대한민국이 화제가 됨"
        },
        {
            "role":"user",
            "content":"그 나라가 화제가 되었던 이유를 자세히 설명해줘"
        },
    ]
)

print(response.output_text)
```

#### Token 관리

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-nano",
    instructions="""
    You are a backend API assistant.
    - Answer concisely and accurately.
    - If unsure, say you are unsure.
    - Do not hallucinate APIs or facts.
    - Prefer bullet points over long prose.
    """,
    input="Are semicolons optional in JavaScript?",
    # max_output_tokens=100
)

print(response.output_text)
```

```python
# 토큰 확인
response.usage
# input_tokens=115
# output_tokens=2147
# total_tokens=2262
total_bill = response.usage.input_tokens / 1000000 * 0.05 + response.usage.output_tokens / 1000000 * 0.4
print(f"{total_bill}")
print(f"{total_bill * 1470} 원")
```

#### 파라메터
- temperature=1 (기본값)
- 기본값보다 올릴 수록 더 창의적이고 예측하기 어려운 형태로 생성
- 기본값보다 낮출수록 일관적이고 논리적인 형태로 생성
- 일반적으로는 0.7 ~ 1.0 사이 선택

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-4o-mini",
    instructions="""
   너는 재밌는 이야기를 만드는 소설가야
    """,
    input="나에게 이야기 해줘",
    temperature=2
)

print(response.output_text)
```
