```python
name = ['Kime', "Park", 'Lee']
try:
    z = 'Lee'
    x = name.index(z)
except:
    print("이름이 없습니다")
else:
    print("이름을 찾았습니다 ", x)
    
```

```python
# 예외 직접 발생 : 자바 => throw new Exception()
# raise

try:
    a = '333'
    if a == 'Kim':
        print('OK')
    else:
        raise ValueError
except ValueError:
    print("에러")
else:
    print(a)
```
