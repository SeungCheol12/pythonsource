#### 정규표현식
- 정규식 패턴은 동일

```python
data = """ park 800905-1049118
kim 700905-1059118
"""

# 뒷자리를 ****** 로 출력

# 엔터를 기준으로 문자열 분리
words = data.split("\n")
print(words)
# 결과를 담을 리스트 생성
result = []
# 공백을 기준으로 문자열 분리
for line in words:
    word_result = []
    for word in line.strip().split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + "-" + "******"
        word_result.append(word)
    # print(word_result)
    print(" ".join(word_result))
    # result.append(" ".join(word_result))

# 리스트를 문자열로
# print("\n".join(result))
```
