```python
class Car:
    color = ""
    speed = 0

    def upSpeed(self, value):
        self.speed += value
    
    def downSpeed(self, value):
        self.speed -= value
    def __str__(self): # == toString()
        return self.color + ":" + str(self.speed)
    
```

```python
car1 = Car()
car1.color = "Red"
car1.speed = 0

car2 = Car()
car2.color = "Black"
car2.speed = 0

car1.upSpeed(80)
car2.upSpeed(100)

print(car1)
print(car2)
```

```python

class SelfTest:
    num = 15

    @staticmethod
    def func1(a, b): # static 메소드() : 변수(클래스변수, self) 사용 안함
        return a + b
    def func2(self): # 인스턴스 메소드
        print("function2 called")
    @classmethod
    def func3(cls): # 클래스 메소드
        print(f"class method {cls.num}")
```

```python
test = SelfTest()
test.func2()

SelfTest.func3()
print(SelfTest.func1(3, 5))
```

```python
# 생성자 오버로딩 없음
# 기본값을 사용해서 오버로딩과 유사한 효과
class UserInfo3:
    def __init__(self, name, age, email=None): # == 생성자
        self.name = name # 인스턴스 변수
        self.age = age
        self.email = email
    def user_info(self): # self == this
        print("메소드 실행")
    def __str__(self):
        return f"name : {self.name} age : {self.age} email : {self.email}"
```

```python
# email 필수 입력 요수가 아닐 때
user1 = UserInfo3("홍길동", 30)
user2 = UserInfo3("성춘향", 40, "user@naver.com")

print(user1)
print(user2)
```

```python
class Car:
    def __init__(self, color="black", speed=0):
        self.color = color
        self.speed = speed

    def upSpeed(self, value):
        self.speed += value
    
    def downSpeed(self, value):
        self.speed -= value
    def __str__(self): # == toString()
        return self.color + ":" + str(self.speed)
```

```python
car1 = Car()
print(car1)
car2 = Car(color="white")
print(car2)
car3 = Car(color="black", speed=100)
print(car3)
```

```python
class Parent():
    def __init__(self):
        self.value = "테스트"

    def test(self):
        print("Parent test() 호출")
class Child(Parent): # 상속
    def __init__(self):
        Parent.__init__(self) # == super() 부모의 인스턴스 변수 호출 여부
        print("Child 클래스의 __init__ 호출")
```

```python
parent = Parent()
parent.test()

child = Child()
child.test()
print(child.value) # => 부모의 인스턴스 변수 호출
```

```python
class AudioVisual:
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume
    
    def switch(self, onOff):
        self.power = onOff
    
    def setVolume(self, volume):
        self.volume = volume

class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)
    
    def tune(self):
        str = "LA la la...." if self.power else "turn it on" # 삼항연산자 대체
        print(str)

class Tv(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        str = "have fun !!" if self.power else "switch on"
        print(str)
```

```python
obj1 = Tv(False, 12, 40)
obj1.switch(True)
obj1.watch()

obj2 = Audio(False, 15)
obj2.setVolume(6)
obj2.tune()
```

```python
class Parent():
    def __init__(self):
        self.value = "테스트"

    def func1(self):
        print("Parent func1() 호출")
    def func2(self):
        print("Parent func2() 호출")
class Child(Parent): # 상속
    def __init__(self):
        Parent.__init__(self) # == super() 부모의 인스턴스 변수 호출 여부
        print("Child 클래스의 __init__ 호출")
    def func1(self):
        print("Parent func1() 재정의")
    def func3(self):
        print("Child func3() 호출")
        
```

```python
parent, child = Parent(), Child()
parent.func1()
parent.func2()

child.func1()
child.func2()
child.func3()
```

```python
# private 키워드는 없음
# __변수 : private 으로 취급됨

import math

class Circle():
    def __init__(self, radius):
        self.__radius = radius
    def getcirc(self):
        return 2 * math.pi * self.__radius
    def getarea(self):
        return math.pi * (self.__radius ** 2)
    def getradius(self):
        return self.__radius
    def setradius(self, radius):
        self.__radius = radius
```

```python
circle = Circle(10)
print(f"둘레 {circle.getcirc()}")
print(f"면적 {circle.getarea()}")
circle.setradius(11)
circle.getradius()
```
