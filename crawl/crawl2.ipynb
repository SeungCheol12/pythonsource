```python
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import requests
import csv
import pymysql
from pymongo import MongoClient
from bson import ObjectId
```

```python
url = "https://openapi.naver.com/v1/search/book.json"
hearders = {
    "X-Naver-Client-Id":"qtOFmvI8ooeSzMWq0fmp",
    "X-Naver-Client-Secret":"j4n25_zV7a"
}
# csv 저장
books = []

# sql 저장
books_sql = []

# mongo 저장
books_mongo = []
start = 1
no = 1
with requests.Session() as s:
        for idx in range(3):
                start_num = start + (idx * 100)
                param = {
                    "query":"해리포터",
                    "display":"100",
                    "start":str(start_num)
                }
                res = s.get(url, headers=hearders, params= param)
                contents = res.json() # python json ==> dict
                for book in contents['items']:
                        # print(books['title'], books['author'], books['discount'], books['publisher'])
                        books.append([no, book['title'], book['author'], book['discount'], book['publisher']])
                        books_sql.append([book['title'], book['author'], book['discount'], book['publisher']])
                        book_m = {
                                "title":book['title'],
                                "author": book['author'],
                                "discount": book['discount'],
                                "publisher":book['publisher']
                        }
                        books_mongo.append(book_m)
                        no += 1
```

```python
with open("./data/books.csv", "w", newline="", encoding="utf-8") as fw:
    wt = csv.writer(fw)

    for v in books:
        wt.writerow(v)
```
