```python
# 주차장 프로그램
# 5대만 주차 가능 (차례대로 입차 / 출차는 반대 순서)
parking_lot = []
top, car_name = 0, "A"
while True:
    no = int(input("[1] 자동차 넣기 | [2] 자동차 빼기 | [3] 종료"))

    if no <= 3:
        if no == 1:
            if top >= 5:
                print("주차장 꽉 찼음")
            else:
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 들어감. 주차장 상태 ==> {parking_lot}")

                top += 1
                # car_name 새로 구하기
                car_name = chr(ord(car_name) + 1)

        elif no == 2:
              if top > 0:
                # 마지막 꺼내기 : pop()
                outCar = parking_lot.pop()
                print(f"{outCar} 자동차 빠짐. 주차장 상태 ==> {parking_lot}")

                top -= 1
                # car_name 새로 구하기
                car_name = chr(ord(car_name) - 1)
              else:
                print("나갈 자동차가 없다")
        elif no == 3:
            print("프로그램 종료")
            break
    else:
        print("매뉴 번호를 확인해 주세요")
```

### set : 집합
- 중복 불가
- 순서가 없음 : 인덱싱 없음

```python
s1 = set()
s2 = set([1,2,3])
s3 = set([1,4,5,6,6])
s4 = set("Hello")
```

```python
print(s1)
print(s2)
print(s3)
print(s4)
```

```python
# print(s2[2]) 인덱스를 이용하고 싶다면?
t1 = tuple(s2)
t1[2]
list1 = list(s2)
list1[2]
```

```python
# 교집합, 합집합, 차집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1 & s2)
print(s1.intersection(s2))

# 합집합
print(s1 | s2)
print(s1.union(s2))

# 차집합
print(s1 - s2)
print(s2 - s1)
print(s1.difference(s2))

# 대칭 차집합
print(s1 ^ s2)
print(s1.symmetric_difference(s2))
```
