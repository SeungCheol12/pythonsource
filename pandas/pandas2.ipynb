```python
# a 칼럼에 5 가 들어있는지 확인
df[df.a.isin([5])]
```

```python
df = pd.DataFrame({"yyyy-mm-dd":['2005-09-08', '2007-10-05', '2012-12-20']})
df
```

```python
def extract_year(value):
    return value.split("-")[0]
```

```python
# year 컬럼 생성
df['year'] = df['yyyy-mm-dd'].apply(extract_year)
df
```

```python
def get_age(year, curr_year):
    return curr_year -int(year)
```

```python
df.dtypes
```

```python
# age 칼럼 생성
df['age'] =  df['year'].apply(get_age, curr_year = 2026)
df
```

```python
df.dtypes
```
