```python
# 1차원 벡터 생성
array1 = np.array([1, 2, 3])
array1
```

```python
print(type(array1))
print(array1.shape) 
```

```python
# 2차원
array2 = np.array([[1, 2, 3], [4, 5, 6]])
array2
```

```python
# 파이썬 리스트
list1 = [1, 2, 'a', 'b', True]
list1
```

```python
array3 = np.array([1, 2, 3.1])
print(array3.dtype)
print(type(array3))
```

```python
array4 = np.array(list1)
print(array4.dtype)
```
