```python
import time
# start = time.time()
# 루프문 문제내기
# time.sleep(3)

# end = time.time()
# print("걸린시간", end-start)
from random import random, uniform, randrange, choice, shuffle, sample
import pymysql
```

```python
# 데이터 베이스 연동
conn = pymysql.connect(host="localhost", user="javadb", password='12345', database="springdb", charset="utf8mb4")
        
# 테이블 생성
with conn.cursor() as cursor:
    sql = "create table if not exists recodes(id int primary key auto_increment, correct_answer int," \
    "recode decimal(10, 2), regdate date default (curdate()))"
    cursor.execute(sql)
```

```python
# list1 = []
with open("./data/word.txt", "r", encoding="utf-8") as fr:
    # for c in fr:
    #     list1.append(c.strip())
    list1 = [line.strip() for line in fr if line.strip()]
    input("게임 시작 엔터키 입력")
    start = time.time()
    n = 1
    corr_cnt = 0
    while n <= 5:
        shuffle(list1)
        q = choice(list1)

        print()
        print(f"** Question #{n}")
        print(q) # 문제 보여주기
        x = input()
        n += 1

        # 문제와 사용자 입력값 일치 여부 확인
        if  x == q:
            corr_cnt += 1
            print("Pass!!")
        else:
            print("Wrong!!")
    end = time.time()

    # 걸린시간
    et = end - start
    et = f"{et:.2f}"
   
    print("*** 게임결과 ***")
    print(f"게임시간 {et} 초, 정답 개수 : {corr_cnt}")
    if corr_cnt > 3:
        print("합격")
    else:
        print("불합격")

# 데이터 삽입
with conn.cursor() as cursor:
    sql = "insert into recodes(correct_answer, recode) values(%s, %s) "
    
    cursor.execute(sql, (corr_cnt, et))
conn.commit()
```
