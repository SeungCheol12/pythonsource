## Pandas
- 테이블 형태의 데이터를 다루는데 사용하는 데이터 분석 패키지
- 구조 - Series, DataFrame

```python
import pandas as pd
```

```python
series = pd.Series([70, 60, 90])
series
```

```python
series = pd.Series({"a":1, "b":2, "c":3})
series
```

```python
series = pd.Series([70, 60, 90], index=["국어", "수학", "영어"])
series
```

```python
# 조회

print(series.국어)
print(series.index)
print(series.values)
print(series.shape) # (3, ) => 튜플
print(series.head(1))
print(series.tail(1))
```

```python
# 수정
# index 수정
series.index = ["미술", "음악", "체육"]
series

# dtype: int64 => int 64비트(8byte) 표현
# java : int 4byte, long 8byte

# 타입변경
series = series.astype("int32")
series
```

```python
# 제거
del series["미술"]
series
```
