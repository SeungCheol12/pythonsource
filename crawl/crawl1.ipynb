#### 크롤링
- 크롤링 : 주기적으로 / 무작위로 데이터 수집 => 데이터 양을 우선으로 수집
    - robots.txt : 크롤링 규칙에 대해 각 사이트마다 지정된 정보 => 각 사이트 주소 뒤에 /robots.txt 작성하면 출력
        - https://www.naver.com/robots.txt => \
        User-agent: * : 모든 크롤러 \
            Disallow: /	: 사이트 전체 크롤링은 금지 \
            Allow : /$ 	: https://www.naver.com/ 만 허용 \
            Allow : /.well-known/privacy-sandbox-attestations.json : 이 파일도 허용 
- 스크랩핑 : 특정 카테고리 지정해서 일부 가져오기(크롤링 보다 작은 개념)

#### 2. requests 라이브러리
- urllib.request 모듈과 같은 역할
- 디코딩을 자동으로 처리, json 처리가 간편
- get, post, delete 메소드 제공
- 세션 적용 가능

```python
# !pip install requests
```

```python
import requests
```

```python
url = "https://naver.com"
res = requests.get(url)
print(res)
print(res.status_code)
print(res.headers.get('content-type'))
print(res.text)
```

```python
url = "https://www.encar.com/"
res = requests.get(url)
print(res)
print(res.headers.get('content-type'))
```

```python
url = "https://naver.com"
s= requests.Session()
r = s.get(url)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python
# put, post, delete => 

url = "https://httpbin.org"

s= requests.Session()

data = {"name":"hong"}

r = s.post(f"{url}/post", data=data)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python

s= requests.Session()


r = s.delete(f"{url}/delete")
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python

s= requests.Session()

data = {"name":"hong"}

r = s.put(f"{url}/put", data=data)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python

s= requests.Session()

param = {"name":"hong"}

r = s.get(f"{url}/get", params=param)
print(r.ok)
print(r.text)

# 세션 종료
# s.close()
```

```python
url = "https://jsonplaceholder.typicode.com/todos/1"
with requests.Session() as s:
    res = s.get(url)
    contents = res.json() # python json ==> dict
    for k,v in contents.items():
        print(k, v)
```
