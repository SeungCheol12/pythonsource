```python
# 사칙연산기 (숫자 2, 연산자 => +, -, *, / , // , %)

num1 = int(input("첫번째 : "))
op = input("연산")
num2 = int(input("두번째 : "))

result = 0
if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/":
    result = num1 / num2
elif op == "//":
    result = num1 // num2
elif op == "%":
    result = num1 % num2

print(f"{num1} {op} {num2} = {result}")
```
