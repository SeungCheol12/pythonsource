```python
# sort({필드명 : 1}) : 오름차순 or # sort(필드명 , -1) : 내림차순
# find({},{"title":1, "imdb.rating":1, "_id":0}) : 전체 조회 하되, 컬럼명은 지정한 것만 보여주기
docs = db.movies.find({},{"title":1, "imdb.rating":1, "_id":0}).sort({"imdb.rating":1})
for doc in docs:
    print(doc)
```

```python
docs = db.movies.find({"imdb.rating":{"$gte":8}},{"title":1, "_id":0 , "imdb":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# imdb 8점이상이고, 2000년 이후인 영화 조회 (제목, imdb, year)
docs = db.movies.find({"imdb.rating":{"$gte":8}, "year":{"$gte":2000}},{"title":1, "_id":0, "imdb":1, "year":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 제목에 love(대소문자 구분없이)
docs = db.movies.find({"title":{"$regex":"love", "$options":"i"}},{"title":1, "_id":0, "imdb":1, "year":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 제작연도가 1920년 이전인 영화 개수
docs = db.movies.find({ "year":{"$lte":1920}},{"title":1, "_id":0, "imdb":1, "year":1})
len(docs.to_list())
# 전체 문서 수 => db.movies.count_documents({})
# db.movies.count_documents({"year":{"$lte":1920}})
```

```python
# 장르가 Comedy, Romance 둘 다 포함된 영화, title, genres 10개만
# in : or
# $all : 모두 를 사용
docs = db.movies.find({"genres":{"$all":["Comedy", "Romance"]}},{"title":1, "_id":0,"genres":1 })
for doc in docs:
    print(doc)
```

```python
# 최신영화 10 개 
docs = db.movies.find({},{"title":1, "_id":0,"genres":1 ,"year":1}).sort({"year":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 영화 제작년도가 1990 ~ 1999 년 사이의 영화 조회
docs = db.movies.find({"year":{"$gte":1990, "$lte":1999}},{"title":1, "_id":0,"genres":1 ,"year":1}).sort({"year":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 모든 장르 보기
print(db.movies.distinct("genres"))
```

```python
# 연도별 영화 개수
rows = db.movies.aggregate([
                    {"$group":{"_id":"$year", "count":{"$sum":1}}},
                    {"$sort":{"_id":-1}},
                    {"$limit":10}

])
for row in rows:
    print(row)
```

```python
# 장르별 영화 개수
rows = db.movies.aggregate([
                    {"$unwind":"$genres"},
                    {"$group":{"_id":"$genres", "count":{"$sum":1}}},
                    {"$sort":{"_id":-1}},
                    {"$limit":10}

])
for row in rows:
    print(row)
```

```python
# imdb 평점 평균이 높은 감독
rows = db.movies.aggregate([
                    {"$unwind":"$directors"},
                    {"$group":{"_id":"$directors", "avgRating":{"$avg":"$imdb.rating"},"count":{"$sum":1}}},
                    {"$match":{"count":{"$gte":3}}},
                    {"$sort":{"avgRating":-1}},
                    {"$limit":10}

])
for row in rows:
    print(row)
```
